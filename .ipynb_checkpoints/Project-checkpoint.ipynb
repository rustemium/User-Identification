### Создадим новые признаки

Сначала создадим признак *number_of_sites*, который будет показывать количество посещенных сайтов за одну сессию

```python
if 'target' in data.columns:
    data['number_of_sites'] = ((data.notna().sum(axis=1)) / 2) - 1
else:
    data['number_of_sites'] = ((data.notna().sum(axis=1)) / 2) - 0.5
data['number_of_sites'] = data['number_of_sites'].astype('int64')
data.head()
```

*session_end* — время конца сессии

*session_length* — длина сессии в секундах

*weekday* — день недели сессии (от 0 до 6)

*start_hour* — час начала сессии (от 0 до 23)

*end_hour* — час конца сессии (от 0 до 23)

*day* — день сессии (от 1 до 31)

*minute* — минута начала сессии (от 0 до 59)

*month* — месяц начала сессии (от 1 до 12)

```python
data[times] = data[times].apply(pd.to_datetime)
data['session_start'] = data[[f"time{i}" for i in range(1,11)]].min(axis=1)
data['session_end'] = data[[f"time{i}" for i in range(1,11)]].max(axis=1)
data['session_time'] = (data['session_end'] - data['session_start']).dt.seconds
data['weekday'] = data['session_start'].dt.dayofweek
data['weekday'] = data['weekday'].astype('category')
data['start_hour'] = data['session_start'].dt.hour
data['start_hour'] = data['start_hour'].astype('category')
data['end_hour'] = data['session_end'].dt.hour
data['end_hour'] = data['end_hour'].astype('category')
data['day'] = data['session_start'].dt.day
data['day'] = data['day'].astype('category')
data['minute'] = data['session_start'].dt.minute
data['month'] = data['session_start'].dt.month
data.head()
```

Создание словаря доменов

```python
def create_site_dict(data, site_dict):
    sites = ['site' + str(i) for i in range(1, 11)]
    site_id = 0
    for site in sites:
        for i in range(len(data[site])):
            if not pd.isna([data[site][i]]):
                if data[site][i] not in site_dict:
                    site_id += 1
                    site_dict[data[site][i]] = site_id
                    data[site][i] = site_id
                else:
                    data[site][i] = site_dict[data[site][i]]
    return site_dict
```

```python
%%time
site_dict = {}
site_dict = create_site_dict(data, site_dict)
```

```python
with open('site_dict.pickle', 'wb') as f:
    pickle.dump(site_dict, f)
```
